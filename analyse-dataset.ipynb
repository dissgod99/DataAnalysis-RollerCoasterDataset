### Step 02: Data Preprocession

* Drop irrelevant columns and rows
* Identify duplicates
* Rename columns
* Feature creation

```python
df = df[['coaster_name', 
    #'Length', 'Speed', 
    'Location', 'Status', 
    #'Opening date',
    #'Type',
    'Manufacturer', 
    #'Height restriction', 'Model', 'Height',
    #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
    #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
    #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
    #'Track layout', 'Fastrack available', 'Soft opening date.1',
    #'Closing date', 'Opened', 'Replaced by', 'Website',
    #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
    #'Single rider line available', 'Restraint Style',
    #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
    'year_introduced', 
    'latitude', 'longitude', 'Type_Main',
    'opening_date_clean', 
    # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
    'speed_mph', 
    # 'height_value', 'height_unit', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.shape
```

```python
df["opening_date_clean"] = pd.to_datetime(df["opening_date_clean"])
```

```python
df["opening_date_clean"]
```

```python
# Rename columns
new_columns = {
    "coaster_name": "Coaster_Name",
    "year_introduced": "Year_Introduced",
    "latitude": "Latitude",
    "longitude": "Longitude",
    "opening_date_clean": "Opening_Date",
    "speed_mph": "Speed_mph",
    "height_ft": "Height_ft",
    "Inversions_clean": "Inversions",
    "Gforce_clean": "Gforce"
}
df = df.rename(columns=new_columns)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.duplicated(subset=["Coaster_Name"]).value_counts()
```

```python
#checking an example duplicate
df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
df.duplicated(subset=["Coaster_Name", "Location", "Opening_Date"]).value_counts()
```

```python
df = df.loc[~df.duplicated(subset=["Coaster_Name", "Location", "Opening_Date"])].reset_index(drop=True).copy()
```

### Step 03: Feature Understanding
* Plotting Feature Distributions
    * Histogram
    * KDE
    * Boxplot

```python
df["Year_Introduced"].value_counts()
```

```python
ax = df["Year_Introduced"].value_counts() \
    .head(10) \
    .plot(kind="bar", title="Top Years Coasters Introduced")

ax.set_xlabel("Year Introduced")
ax.set_ylabel("Count")
```

```python
ax = df["Speed_mph"].plot(kind="hist", 
                          bins=20, 
                          title="Coaster Speed (mph)")
ax.set_xlabel("Speed (mph)")
```

```python
ax = df["Speed_mph"].plot(kind="kde",
                          title="Coaster Speed (mph)", color="b")
ax.set_xlabel("Speed (mph)")
```

```python
sns.histplot(data=df["Speed_mph"], kde=True)
```
